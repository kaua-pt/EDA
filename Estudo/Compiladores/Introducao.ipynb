Compilador: lê um programa em uma linguagem e o transfere para outra linguagem(linguagem alvo),o processo de compilação deve relatar erros, normalmente, as linguagens fonte(que sera traduzida) é uma linguagem de alto nível, temos vários tipos de compiladores

Antes de passar pelo compilador, um programa pode ser pre-processado(em c utilizamos include ou define) e após a compilação, a linguagem pode exigir processos adicionais



A compilação contém a parte de análise e síntese.
Analise: divide o programa e organiza-o de uma forma intermediária, que temos uma árvore sintática em que cada nó representa uma operação e cada filho um operando

Síntese: constroi o programa



O ramo a esquerda sofre a operação

#### Analise

Contém 3 fases:
1- Linear,lexica ou esquadrinhamento -> fluxo de caracteres do programa lido está agrupado em tokens com significado.Imagine F= 1.8 * C + 32 , o programa separa os tokens F,=,1.8, etc

2- Hierarquica,sintatica ou Gramatical -> Tokens ordenados hierarquicamente com significado.

3- Semântica -> Componentes do programa se conectam de forma significativa
